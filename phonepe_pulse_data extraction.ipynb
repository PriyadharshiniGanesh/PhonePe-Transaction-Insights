```python
import json
```

```python
clm={'State':[], 'Year':[],'Quater':[],'User_id':[], 'User_brand':[], 'User_count':[], 'User_percentage':[]}

for i in Agg_user_list:
    p_i=os.path.join(path1, i)
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=os.path.join(p_i, j)
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=os.path.join(p_j, k)
            try:
                with open(p_k, 'r') as Data:
                    D = json.load(Data)
                    # Skip if 'data' is missing
                    if 'data' not in D:
                        continue
                    # Extract aggregated data (registeredUsers)
                    registered_users = D['data'].get('aggregated', {}).get('registeredUsers', 0)
                    # Extract usersByDevice (must be a list)
                    users_by_device = D['data'].get('usersByDevice', [])
                    for device in users_by_device:
                        brand = device.get('brand', 'Unknown')
                        count = device.get('count', 0)
                        percentage = device.get('percentage', 0.0)
                        clm['User_id'].append(registered_users)
                        clm['User_brand'].append(brand)
                        clm['User_count'].append(count)
                        clm['User_percentage'].append(percentage)
                        clm['State'].append(i)
                        clm['Year'].append(j)
                        clm['Quater'].append(int(k.strip('.json')))
            except Exception as e:
                print(f"Error processing {p_k}: {e}")
                continue

# Create DataFrame
Agg_User = pd.DataFrame(clm)
```

```python
path7=r"C:\\Users\vicky\Documents\PHONEPAY\data\data\top\user\country\india\state\\"
top_user_list=os.listdir(path7)
```

```python
clm = {'State': [], 'Year': [], 'Quater': [], 'District': [], 'User_id': [], 'pincode': []}

for i in top_user_list:
    p_i = os.path.join(path7, i) + "/"
    try:
        Agg_yr = os.listdir(p_i)
    except FileNotFoundError:
        print(f"Directory not found: {p_i}")
        continue
    for j in Agg_yr:
        p_j = os.path.join(p_i, j) + "/"       
        try:
            Agg_yr_list = os.listdir(p_j)
        except FileNotFoundError:
            print(f"Directory not found: {p_j}")
            continue      
        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)
            
            try:
                with open(p_k, 'r') as Data:
                    try:
                        D = json.load(Data)
                        
                        # Process districts
                        if D and D.get('data') and D['data'].get('districts'):
                            for district in D['data']['districts']:
                                # Process pincodes for each district
                                if D['data'].get('pincodes'):
                                    for pincode in D['data']['pincodes']:
                                        clm['District'].append(district.get('name'))
                                        clm['pincode'].append(pincode.get('name'))
                                        clm['User_id'].append(district.get('registeredUsers', 0))  # Assuming this is what you want
                                        clm['State'].append(i)
                                        clm['Year'].append(j)
                                        clm['Quater'].append(int(k.strip('.json')))
                                       
                                else:
                                    # If no pincodes, just add district info
                                    clm['District'].append(district.get('name'))
                                    clm['pincode'].append(None)
                                    clm['User_id'].append(district.get('registeredUsers', 0))
                                    clm['State'].append(i)
                                    clm['Year'].append(j)
                                    clm['Quater'].append(int(k.strip('.json')))
                                  
                        else:
                            print(f"No districts data in file: {p_k}")
                            
                    except json.JSONDecodeError:
                        print(f"Invalid JSON in file: {p_k}")
                        
            except IOError:
                print(f"Could not read file: {p_k}")

# Create DataFrame
Top_user = pd.DataFrame(clm)
```
